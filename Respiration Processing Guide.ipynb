```python
# Respiration Processing Guide
__author__ = 'Catia'

from pylab import *
from numpy import *
from scipy import *

%pylab inline
import novainstrumentation as ni
ni.figtools.pylabconfig()
```

```python
def respFilt(signal, fs, step,  win):
    signal = signal[:]*-1
    signal = signal - mean(signal)
    signal = signal - ni.smooth(signal, win)
    signal = signal/std(signal)
    signal = ni.smooth(signal, win*2/30)
    t = arange(len(signal))/fs*step
    return [t, signal]


def getExp(signal, th_exp):
    p_exp = ni.peakdelta(signal, th_exp)[0][:, 0].astype('int')

    return p_exp


def getInsp(signal, th_insp):
    p_ins = ni.peakdelta(diff(diff(signal) > th_insp) > 0.5, th_insp)[0][:, 0].astype('int')
    if p_ins[0] < p_exp[0] < p_ins[1]:
        p_ins = p_ins[1::2]
    else:
        p_ins = p_ins[0::2]
    return p_ins


def respmw(signal, p_exp, th_insp, lowerBounder, higherBounder):
    r_w = ni.waves(signal, p_exp, lowerBounder, higherBounder)
    m_w = ni.meanwave(r_w)
    s_w = std(r_w, 0)

    icycle_start = find(diff(m_w) > th_insp)[0]  # Detect inspiration
    icycle_expiratory = argmax(m_w)  # Detect expiration
    # Detect expiration pause
    icycle_exp_pause = find(m_w[icycle_expiratory:] < m_w[icycle_start])[0] + icycle_expiratory
    icycle_end = find(m_w < m_w[icycle_start])[-1]  # Detect end

    # xAxis in percentage
    m = 100./(icycle_end-icycle_start)
    b = -icycle_start*m

    pcycle = arange(len(m_w))*m+b

    pcycle_start = 0
    pcycle_expiratory = pcycle[icycle_expiratory]
    pcycle_exp_pause = pcycle[icycle_exp_pause]
    pcycle_end = 100

    figure(figsize=(20, 6))

    plot(pcycle, m_w, 'k', lw=2)  # mean wave
    plot(pcycle, m_w+s_w, '-.', color='0.5')  # mean wave + std
    plot(pcycle, m_w-s_w, '-.', color='0.5')  # mean wave - std
    axis([-20, 120, m_w[icycle_exp_pause]-s_w[icycle_exp_pause]-0.5*m_w[icycle_expiratory], m_w[icycle_expiratory] +
          s_w[icycle_expiratory]+0.3*m_w[icycle_expiratory]])

    # Parameters definition
    for t_, p_, i_ in [['inspiration start', pcycle_start, icycle_start],
                       ['expiration start', pcycle_expiratory, icycle_expiratory],
                       ['expiratory \n pause', pcycle_exp_pause, icycle_exp_pause],
                       ['cycle end', pcycle_end, icycle_end]]:
        annotate(t_, xy=(p_, m_w[i_]), xytext = (p_ + 5, m_w[icycle_expiratory]+0.1*m_w[icycle_expiratory]),
                 size='large', arrowprops = dict(arrowstyle="->", connectionstyle="angle,angleA=0, angleB=90, rad=10"))

    xlabel('% Cycle')
    ylabel('amplitude (V)')
    title('Respiration Mean Wave')
    show()
```

```python
# OPEN SIGNAL 

signal = loadtxt('v2_simple breath_20150316_11h41m00s.txt')
fs = 1000.
step = 10
resp = signal[30000:50000:step, 1]
plot(resp)
```

```python
# SIGNAL FILTRATION

t, respf = respFilt(resp, fs, step, 800)
```

```python
# DETECT EXPIRATION

TH_EXPIRATION = 1  # delta do detect the beggining of expiration
p_exp = getExp(respf, TH_EXPIRATION)
```

```python
# DETECT INSPIRATION

TH_CYCLE_START = 0.025  # threshold to detect the beggining of inspiration  
p_ins = getInsp(respf, TH_CYCLE_START)
```

```python
# FILTERED SIGNAL
# INSPIRATION MARK
# EXPIRATION MARK

figure(figsize=(20, 3))
plot(t, respf, '0.3')
vlines(t[p_exp], respf[argmin(respf)], respf[argmax(respf)], 'g')
vlines(t[p_ins], respf[argmin(respf)], respf[argmax(respf)], 'y')
xlabel('time (s)')
ylabel('amplitude (V)')
title('Respiration wave - expiration (green) and inspiration (yellow)')
axis([0, t[-1], respf[argmin(respf)]-0.5, respf[argmax(respf)]+0.5])
```

```python
# MEAN WAVE

respmw(respf, p_exp, TH_CYCLE_START, -120, 220)
```
